```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset('mpg')
```

```python
df
```

```python
sns.jointplot(x='mpg', y='acceleration', data = df)
```

The univariatn of this data is this joint plot because it compares two different variables in one graph

```python
sns.jointplot(x='mpg', y='acceleration', data = df, kind='hex')
```

```python
sns.jointplot(x='mpg', y='acceleration', data = df, kind ='kde')
```

```python
sns.jointplot(x='mpg', y='acceleration', data = df, kind ='scatter')
```

```python
sns.jointplot(x='mpg', y='acceleration', data = df, kind ='reg')
```

```python
sns.jointplot(x='mpg', y='acceleration', data = df, kind ='resid')
```
